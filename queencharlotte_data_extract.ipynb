```python
# Search for tweets with specific keywords (excluding retweets)
keyword1 = "QueenCharlotte"

query = f"{keyword1} -filter:retweets"
tweets = tweepy.Cursor(api.search_tweets, q=query, lang='en', tweet_mode="extended").items(100000)

# Create an empty DataFrame
df = pd.DataFrame(columns=["Text", "User", "Description", "Timestamp", "Location", "Likes", "Hashtags"])


# Populate the DataFrame with fetched tweet data
for tweet in tweets:
    hashtags = [tag["text"] for tag in tweet.entities["hashtags"]]

    df = df.append({
        "Text": tweet.full_text,
        "User": tweet.user.screen_name,
        "Description": tweet.user.description,
        "Timestamp": tweet.created_at,
        "Location": tweet.user.location,
        "Likes": tweet.favorite_count,
        "Hashtags": ", ".join(hashtags)
    }, ignore_index=True)

# Print the DataFrame
print(df.shape)
```
